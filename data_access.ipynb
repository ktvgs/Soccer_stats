```python
events=pd.DataFrame()
```

```python
c=0
for i in matches['match_id']:
    new=pd.DataFrame(sb.events(match_id=i))
    events=events.append(new,ignore_index=True)
    c+=1
    if c>10:
        break
```

```python
events.shape
```

```python
events['type'].unique()
```

```python
selected_types = ['Pass', 'Shot', 'Duel', 'Dribble']

# Filtering the DataFrame
filtered_events = events[events['type'].isin(selected_types)]
```
