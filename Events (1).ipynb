```python
barca=[]
for i in np.array(list(playf[playf['team']=='Barcelona']['location']))-np.array(list(playf[playf['team']=='Barcelona']['pass_end_location'])):
  if i[0]>0:
    barca.append('forward')
  elif i[0]<0:
    barca.append('backward')
  else:
    barca.append('side')
```

```python
plt.hist(barca)
```

```python
madrid=[]
for i in np.array(list(playf[playf['team']!='Barcelona']['location']))-np.array(list(playf[playf['team']!='Barcelona']['pass_end_location'])):
  if i[0]>0:
    madrid.append('forward')
  elif i[0]<0:
    madrid.append('backward')
  else:
    madrid.append('side')
```

```python
barca_team={}
for i,idx in playf[playf['team']=='Barcelona'].loc[:,['player','pass_recipient']].iterrows():
  if (idx['player'],idx['pass_recipient']) not in barca_team:
    barca_team[(idx['player'],idx['pass_recipient'])]=1
  else:
    barca_team[(idx['player'],idx['pass_recipient'])]+=1
```
